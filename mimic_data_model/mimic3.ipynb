```python
import pandas as pd
import os

import ipywidgets as widgets
from IPython.display import display, clear_output
```

```python
filepath = "../rawdata/mimic-iii-demo/"
csvs = [csv for csv in os.listdir(filepath) if csv[-4:] == ".csv"]

csv_files = []
for csv in csvs:
    csv_files.append(os.path.join(filepath,csv))
```

```python

# Define the function to read the selected file
def read_file(filename):
    df = pd.read_csv(filename)
    display(df.head())
    
    
    
# Define the function to update the dataframe when the dropdown value changes
def on_change(change):
    if change['type'] == 'change' and change['name'] == 'value':
        # clear screen, 
        clear_output(wait=False)
        display(dropdown)
        read_file(change['new'])
    
```

```python
# Define the dropdown widget
dropdown = widgets.Dropdown(
    options=csv_files,
    value=csv_files[0],
    description='Select file:'
)


# Display the dropdown widget and initial dataframe
display(dropdown)
read_file(dropdown.value)

# Call on_change() function when dropdown value changes
dropdown.observe(on_change)
```

```python
csv_files.sort(key = lambda x: os.path.basename(x))
for csv_file in csv_files:
    print(csv_file)
```
