# EHR data warehouse design

The next step is to develop a way to collectively understand how to merge a series of csv files from mimic 3 and 4 into one.


```python
import pandas as pd
import os
import dataprofiler as dp


from util.helpers import find_csvs
import json
import pprint
```

```python
def get_table_schema(filepath):
    # read in the file and do something about it
    filepathes = find_csvs(data_direcotry=filepath)
    tables = [[os.path.splitext(os.path.basename(filepath))[0],filepath] for filepath in filepathes]
    
    res = dict()
    for i, table in enumerate(tables):    
        profile = dp.Profiler(pd.read_csv(table[1]))
        report = profile.report(report_options = {'output_format':'pretty'})
        
        df = pd.DataFrame.from_dict(report["data_stats"])
        my_dict = df.set_index('column_name')['data_type'].to_dict()
        my_dict.update({"filepath":str(table[1])})
        res[str(table[0])] = my_dict
    
    return res
```

```python


mimic_3 = get_table_schema(filepath = "../basic_filtered_data/mimic-iii-demo/")
    
```
